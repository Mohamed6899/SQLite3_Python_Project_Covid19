```python
# Importing The Libraries
import sqlite3 , pandas as pd 
```

```python
# Creating The Connection
con = sqlite3.connect('covid19.db')
cur = con.cursor()
```

```python
# Importing The Data Into SQL using pandas 

CovidDeath = pd.read_csv(r"D:\Datasets\CovidDeath.csv")
CovidVaccinations = pd.read_csv(r"D:\Datasets\CovidVaccinations.csv")
cur.execute('''DROP TABLE IF EXISTS CovidDeath''')
cur.execute('''DROP TABLE IF EXISTS CovidVaccinations''')
CovidDeath.to_sql('CovidDeath' , con)
CovidVaccinations.to_sql('CovidVaccinations' , con)
```

```python
# Exploring the Data 

statment = '''SELECT 
location ,
date , 
total_cases ,
total_deaths ,
new_cases ,
population 
FROM CovidDeath
order by 1 , 2;'''
cur.execute(statment)
result = cur.fetchall()
result
```

```python
# Looking at Total Cases vs Population

statment = '''SELECT 
	location , 
	date , 
	population , 
	total_cases  , 
	total_cases / NULLIF(population, 0) * 100 AS PercentPopulationInfected
		FROM CovidDeath
	where location  like 'eg%' 
	order by 1 , 2;'''
cur.execute(statment)
result = cur.fetchall()
for row in result:
    print(row)
```

```python
#-- Looking at countries with highest infection rates Compared to Population

statment = '''SELECT 
	location , 
	population , 
	max(total_cases) as HighestInfectionCount  ,
	max(total_cases / NULLIF(population, 0)) * 100 AS PercentPopulationInfected
		FROM CovidDeath
        	Group by location , population
	order by PercentPopulationInfected desc;'''
cur.execute(statment)
result = cur.fetchall()
result
```

```python
# Showing Countries With Highest Death Count Per Population 
statment = '''SELECT 
	location , 
	max(total_deaths) as TotalDeathCount  
		FROM CovidDeath 
--where location  like 'eg%' 
	Group by location
	order by TotalDeathCount desc;'''
cur.execute(statment)
result = cur.fetchall()
result 
```

```python
# Showing Continent  With Highest Death Count Per Population 

statment = '''SELECT 
	Continent  , 
	max(total_deaths) as TotalDeathCount  
		FROM CovidDeath 
--where location  like 'eg%'
	where Continent != ' '
	Group by Continent
	order by TotalDeathCount desc;'''
cur.execute(statment)
result = cur.fetchall()
result 
```

```python
# global Numbers

statment = '''SELECT date , 
	sum(total_cases) as Total_Cases ,
	sum(new_deaths) as Total_Deaths ,
	(sum(new_deaths) / nullif(sum(new_cases) ,0)) *100 as Death_Percentage 
		FROM CovidDeath
--where location  like 'eg%' 
	where Continent IS NOT NULL
	group by date
	order by 1 , 2;'''
cur.execute(statment)
result = cur.fetchall()
result 
```

```python
# Total_Cases , Total_Deaths And The Death Percentage 

statment = '''SELECT 
	sum(total_cases) as Total_Cases ,
	sum(new_deaths) as Total_Deaths ,
	(sum(new_deaths) / nullif(sum(new_cases) ,0)) *100 as Death_Percentage 
		FROM CovidDeath
--where location  like 'eg%' 
	where Continent != ' '
	order by 1 , 2;'''
cur.execute(statment)
result = cur.fetchall()
result
```

```python
# Calculation total population vs vaccinations

statment = '''select 
	dea.continent , 
	dea.location , 
	dea.date , 
	dea.population ,
	vacc.new_vaccinations , 
		sum(vacc.new_vaccinations) over (partition by dea.location order by dea.location ,dea.date ) as RollingPeopleVaccinated 
from CovidDeath dea  
	join CovidVaccinations vacc 
		on dea.location = vacc.location 
		and dea.date = vacc.date 
where dea.continent != ' ' and vacc.new_vaccinations!= 0
order by 1 ,2'''
cur.execute(statment)
output =  cur.fetchall()
output
```

```python
# Temp Table 

cur.execute('drop table if exists percentPopulationVaccinated')
create_table_statement = '''

create table percentPopulationVaccinated
(
	Continent nvarchar(255),
	Location nvarchar (255),
	Date datetime ,
	Population numeric ,
	New_vaccinations numeric,
	Rollingpeoplevaccinated numeric
);

insert into percentPopulationVaccinated
select 
	dea.continent , 
	dea.location , 
	dea.date , 
	dea.population ,
	vacc.new_vaccinations , 
		sum(vacc.new_vaccinations) over (partition by dea.location order by dea.location ,dea.date ) as RollingPeopleVaccinated 
from CovidDeath dea  
	join CovidVaccinations vacc 
		on dea.location = vacc.location 
		and dea.date = vacc.date 
where dea.continent != ' ' and vacc.new_vaccinations!= 0
order by 1 ,2;
'''
cur.executescript(create_table_statement)
con.commit()
```
